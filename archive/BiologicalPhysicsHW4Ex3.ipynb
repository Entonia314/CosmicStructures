```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
d_S = 2  # DNA diameter in nm
D = 40  # Capsid diameter in nm
H = 50  # Capsid height in nm
A = 50  # Persistence length in nm
separation_distance = np.sqrt(3 * d_S / 2)  # Separation distance in nm
```

```python
# Calculate the length of DNA packed in n layers
def packed_dna_length(layers):
    R1 = D / 2  # Radius of the first layer in nm
    total_length = 0
    for n in range(1, layers + 1):
        Rn = R1 - (n - 1) * separation_distance
        length_n = 25 * 2 * np.pi * Rn
        total_length += length_n
    return total_length

# Calculate packed DNA lengths for layers from 1 to 10
layers = np.arange(1, 11)
packed_lengths = [round(packed_dna_length(n),2) for n in layers]
print(packed_lengths)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(layers, packed_lengths, marker='o', linestyle='-', color='b')
plt.title('Total Packed DNA Length vs. Number of Filled Layers')
plt.xlabel('Number of Filled Layers')
plt.ylabel('Total Packed DNA Length (nm)')
plt.grid(True)
plt.xticks(layers)
plt.show()
```

```python
# Calculate the energy of DNA packed in n layers
def packed_dna_energy(layers):
    R1 = D / 2  # Radius of the first layer in nm
    total_length = 0
    total_energy = 0
    for n in range(1, layers + 1):
        Rn = R1 - (n - 1) * separation_distance
        length_n = 25 * 2 * np.pi * Rn
        total_length += length_n
        energy_n = 0.5 * A / (Rn ** 2) * length_n
        total_energy += energy_n
    return total_energy

# Calculate packed DNA energy for layers from 1 to 10
layers = np.arange(1, 11)
packed_dna_energy = [round(packed_dna_energy(n), 2) for n in layers]
print(packed_dna_energy)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(layers, packed_dna_energy, marker='o', linestyle='-', color='b')
plt.title('Total Packed DNA Energy vs. Number of Filled Layers')
plt.xlabel('Number of Filled Layers')
plt.ylabel('Total Packed DNA Energy (kBT)')
plt.grid(True)
plt.xticks(layers)
plt.show()
```
