From Gaussian Random Fields to the Cosmic Web
=========================================

```python
import numpy as np
import matplotlib.pyplot as mplt
import ipywidgets as widgets  # only needed for isosurfaces, comment out if you don't have it installed
import ipyvolume as ipv # only needed for isosurfaces, comment out if you don't have it installed
%matplotlib inline
np.seterr(all='ignore')
```

```python
##################################################################################
N = 128                           # we discretize our volume using N**3 cells
L = 1.0                           # the box size, sets the fundamental mode
```

Generating a white Gaussian random field
========================================

A Gaussian random field with no correlations is a real scalar random field defined for every point of a multidimensional space $S\subset\mathbb{R}^n$. It has a normal distribution at every point $x\in S$. The value at a point $x\in S$ is independent of the value at any other point $y\neq x$.

```python
# create a Gaussian random field of size N**3
delta0 = np.random.normal(size=[N,N,N])
```

All points in a white field are spatially uncorrelated
-----------------------------------------------------------

```python
fig, ax = mplt.subplots(1,2,figsize=[10,5])
fig.suptitle('Gaussian white noise field')
ax[0].imshow(delta0[...,N//2], cmap='jet')
ax[1].hist(delta0.flatten(),bins=60,density=True);
```

Spatial correlation is imposed by the power spectrum
=============================================
A GRF can only have two-point correlations. If it is isotropic and homogeneous, these can only depend on the distance between points.

```python
twopi = 2.0*np.pi
kmin = twopi/L # the fundamental mode of the box
kmax = kmin * N/2     # the Nyquist mode of each linear dimension
k1d = np.fft.fftfreq(N,d=1.0/kmin/N)  # get a 1D vector with all values of 'k'
kv = np.meshgrid(k1d,k1d,k1d) # turn this into 3 3D fields giving kx,ky,kz for each point
kmod = np.sqrt(kv[0]**2+kv[1]**2+kv[2]**2)               # get norm k for each point

# now 'kv' is the k vector, and 'kmod' its modulus at every point
```

```python
alpha = 3.0
norm  = 1.0/twopi**(-1.5)
# the matter power spectrum is the product of primordial spectrum and T**2
def P(k):
    return k**(-alpha)*norm
```

```python
# generate a complex random field
fk = np.random.normal(size=(N,N,N)) + 1j * np.random.normal(size=(N,N,N))
fk = fk * np.sqrt(P(kmod))   # impose power spectrum 
fk[0,0,0] = 0.0     # zero f(k=0) to enforce zero mean

# compute (Fast) Fourier Transform of fk
f = np.real(np.fft.ifftn(fk))

# re-normalise (should be done using filtered variance)
f = f / np.std(f.flatten())
```

The cosmological primordial density field is Gaussian, but with spatial correlations
------------------------------------------------------------------------------------------------

```python
fig, ax = mplt.subplots(1,2,figsize=[10,5])
ax[0].imshow(f[...,N//2], cmap='jet')
ax[1].hist(f.flatten(),bins=60,density=True);
```

Calculate gradient and Hessian of the potential field
==========================================
Let's first remember that gradients become algebraic multiplications in Fourier space, i.e.
$$ \partial_m f \quad\leftrightarrow\quad ik_m \hat{f} $$

The potential and the overdensity field are related through Poisson's equation, with formal solution
$$ \nabla^2 \phi = \delta \quad\leftrightarrow\quad \phi = \nabla^{-2} \delta $$
This symbolic, formal solution is trivially the actual solution on a periodic domain. Then, in Fourier space
$$ -k^2 \hat{\phi} = \hat{\delta} \quad\leftrightarrow\quad \hat{\phi} = -k^{-2}\hat{\delta}\quad\textrm{for}\quad k\neq0.$$

We need the gradient and Hessian field of the potential of our Gaussian density field $\delta$, which are easily evaluated in Fourier space as
$$ D_l = \frac{\partial_l}{|\nabla|} \delta \qquad \leftrightarrow \qquad \hat{D}_l = \frac{i k_l}{|\mathbf{k}|}\hat{\delta}\quad \textrm{for}\quad k\neq0$$ 
and
$$ D^2_{lm} = \frac{\partial_l \partial_m}{\nabla^2} \delta \qquad \leftrightarrow \qquad \hat{D}^2_{lm} = \frac{ k_l k_m}{k^2}\hat{\delta}\quad \textrm{for}\quad k\neq0$$

Finally, we shall also diagonalise the $D^2_{lm}$ field and compute its real eigenvalues $\mu_1\le\mu_2\le\mu_3$ ($D^2$ is symmetric).

```python
# Compute gradient vector and Hessian matrix
fD1 = np.zeros([N,N,N,3],dtype=complex)
fD2 = np.zeros([N,N,N,3,3],dtype=complex)

for i in range(3):
    fD1[...,i] = fk * 1j * kv[i] / kmod
    fD1[0,0,0,i] = 0.0
    
    for j in range(3):
        fD2[...,i,j] = -fk * kv[i] * kv[j] / (-kmod**2)
        fD2[0,0,0,i,j] = 0.0
        
# transform back
D1 = np.real(np.fft.ifftn(fD1,axes=[0,1,2]))
D2 = np.real(np.fft.ifftn(fD2,axes=[0,1,2]))

# also get eigenvalues mu1<mu2<mu3 of Hessian
mu = np.sort(np.linalg.eigvals(D2),axis=-1)
```

```python
fig, ax = mplt.subplots(1,3,figsize=[10,5])
for i in range(3):
    ax[i].imshow(D1[...,N//2,i], cmap='jet')
```

```python
fig, ax = mplt.subplots(3,3,figsize=[10,10])
for i in range(3):
    for j in range(3):
        ax[i,j].imshow(D2[...,N//2,i,j], cmap='jet')
```

The signature of the Hessian field $D^2$
----------------------------------------------

The signature $s(\mathbf{x})$ of $D^2$ at every point $\mathbf{x}$ is defined as the number of its positive eigenvalues $\mu_i$. s is an integer number $0\le s \le 3$.

```python
# also get eigenvalues mu1<mu2<mu3 of Hessian
mu = np.sort(np.linalg.eigvals(D2),axis=-1)
# get signature of eigenvalues
sig = (mu[...,0]>0.0).astype(int)+(mu[...,1]>0.0).astype(int)+(mu[...,2]>0.0).astype(int)

fig, ax = mplt.subplots(1,2,figsize=[10,5])
ax[0].imshow(sig[...,N//4],cmap='jet')
ax[0].set_title('D^2 signature')

ax[1].imshow(f[...,N//4],vmin=-3, vmax=3)
ax[1].set_title('density');
```

```python
print('+++ fraction: ',np.sum(sig==0)/N**3)
print('++- fraction: ',np.sum(sig==1)/N**3)
print('+-- fraction: ',np.sum(sig==2)/N**3)
print('--- fraction: ',np.sum(sig==3)/N**3)
```

## Eigenvalue distributions

```python
import scipy
def p_lambda1( l1, s0 ):
    return (np.sqrt(5)*(20*l1*s0 + np.exp((2*l1**2)/s0**2)*np.sqrt(2*np.pi)*(20*l1**2 - s0**2)*(1 + scipy.special.erf((np.sqrt(2)*l1)/s0)) + \
          3*np.exp((3*l1**2)/(4.*s0**2))*np.sqrt(3*np.pi)*s0**2*(1 + scipy.special.erf((np.sqrt(3)*l1)/(2.*s0)))))/(12.*np.exp((9*l1**2)/(2.*s0**2))*np.pi*s0**3)
def p_lambda3( l3, s0 ):
    return p_lambda1(-l3,s0)
def p_lambda2( l2, s0 ):
    return np.sqrt(15/np.pi)/(2.*s0)*np.exp(-(15*l2**2)/(4.*s0**2))
```

```python
fig, ax = mplt.subplots()

ll = np.linspace(-2.5,2.5,300)

sigma0 = np.std( (mu[...,0]+mu[...,1]+mu[...,2]).flatten() )

ax.plot( ll, p_lambda1(ll,1.0), 'C0', label = '$\\lambda_1$' )
ax.plot( ll, p_lambda2(ll,1.0), 'C1', label = '$\\lambda_2$' )
ax.plot( ll, p_lambda3(ll,1.0), 'C2', label = '$\\lambda_3$' )

ax.hist( mu[...,2].flatten() / sigma0, bins=100, density=True, alpha=0.5, label='mu1');
ax.hist( mu[...,1].flatten() / sigma0, bins=100, density=True, alpha=0.5, label='mu1');
ax.hist( mu[...,0].flatten() / sigma0, bins=100, density=True, alpha=0.5, label='mu1');

ax.set_xlabel('$\\lambda_i / \\sigma_0$')
ax.set_ylabel('$p(\\lambda_i)$')
```
